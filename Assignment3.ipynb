**Task 5 - Checking for Missing values and dealing with them**

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
# Columns for which you want to fill null values with mean
selected_columns = ['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']

# Calculate the mean for the selected columns
mean_values = data[selected_columns].mean()

# Fill null values with the mean for the selected columns
data[selected_columns] = data[selected_columns].fillna(mean_values)

# Calculate the mode of column1
mode_value = data['sex'].mode()[0]

# Fill empty fields in column1 with its mode
data['sex'].fillna(mode_value, inplace=True)

# Print the updated DataFrame
data
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

**Task 6 - Finding the outliers and replacing them**

```python
data
```

```python
Q1 = data[['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']].quantile(0.25)
Q3 = data[['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']].quantile(0.75)
IQR = Q3 - Q1

def replace_outliers(column):
    lower_bound = Q1[column] - 1.5 * IQR[column]
    upper_bound = Q3[column] + 1.5 * IQR[column]
    data[column] = np.where((data[column] < lower_bound) | (data[column] > upper_bound), data[column].median(), data[column])

columns_with_outliers = ['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']
for column in columns_with_outliers:
    replace_outliers(column)
```

**Task 8 - Checking for categorical columns and performing encoding**

```python
le = LabelEncoder()

categorical_columns = ['species', 'island', 'sex']
for column in categorical_columns:
    data[column] = le.fit_transform(data[column])
```

```python
data
```

**Task 9 - Splitting the data into dependent and independent variables**

```python
X = data.drop('species', axis=1)  # Independent variables
y = data['species']  # Dependent variable
```

**Task 10 - Scaling the data**

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

**Task 11 - Splitting the data into training and testing**

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

**Task 12 - Checking the training and testing data shape**

```python
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
plt.plot(y_pred, label='Predicted')
plt.plot(y_test, label='Actual')
plt.legend()
plt.show()
```
